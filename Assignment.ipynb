```python
master1=df1[['ID','SERVICE_type','Unique ID check','Service Name','NET_AMOUNT','Invoice Status','BILLED_DATE','COLLECTED','Completion date']]
master2=df2[['ID','SERVICE Type','Unique ID check','NET_AMOUNT','INVOICESTATUS','BILLED_DATE','COLLECTED','APPROVEDDATETIME']]
master3=df3[['ID','Service Type','Unique ID Check','Service Name','NET_AMOUNT','INVOICESTATUS','Billed Date','Collected Date','APPROVEDDATETIME']]
master1.rename(columns={'SERVICE_type':'Service Type','Unique ID check':'Unique ID','Invoice Status':'INVOICESTATUS','BILLED_DATE':'Billed Date','COLLECTED':'Collected Date','Completion date':'APPROVEDDATETIME'},inplace=True)
master2.rename(columns={'SERVICE Type':'Service Type','Unique ID check':'Unique ID','BILLED_DATE':'Billed Date','COLLECTED':'Collected Date'},inplace=True)
master3.rename(columns={'Unique ID Check':'Unique ID'},inplace=True)
```

```python
frames=[master1,master2,master3]
final_result=pd.concat(frames)
final_result
```

```python
final_result['new_date_col'] = final_result['Billed Date'].dt.date
final_result
```

```python
import numpy as np 
duplicate = final_result.duplicated(['ID','new_date_col','Service Type'],keep=False)
final_result['Unique ID']= np.where(duplicate==True,'D',final_result['Unique ID'])
final_result = final_result.replace('nan', np.nan)
final_result
```

```python

summary1= pd.pivot_table(final_result,index=['new_date_col'],
                    columns=['Service Type'],aggfunc=len)
summary1=summary1['APPROVEDDATETIME']
summary1
```
